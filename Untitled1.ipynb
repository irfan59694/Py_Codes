```python
import numpy as np
import matplotlib.pyplot as plt

mean=0
std_dev=1
sample=1000
data=np.random.normal(mean, std_dev, sample)

#ploting of normal distrubtuion
plt.hist(data, alpha=0.6, bins=30, color='g', density='true')
#In the above, we have generated normal dist. taking mean 0 and standard deviatio 1 (just for example) and then created normal distrbitution of 1000 sample, then created histo and you can check its approx symmterical histo)
#now have look for further info
#now we are going to plot the probability density function (PDF) of the normal distribution.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-(x - mean)**2 / (2 * std_dev**2)) / (np.sqrt(2 * np.pi) * std_dev) #dont worry. its probablity formula
plt.plot(x, p, color='k', linewidth=2)
#modification
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
#Further Briefing_IMP; In the code:We first create an array x of values ranging from the minimum to the maximum of the generated data.
#Then, we compute the corresponding probability density at each value of x using the above formula.
#Finally, we plot these values on the same graph using matplotlib.pyplot.plot.
#The purpose of plotting the probability density function alongside the histogram is to visually compare the theoretical distribution (given by the PDF) with the empirical distribution (given by the histogram of the generated data). This helps us understand how well the data fits the theoretical normal distribution. In most cases, with a sufficiently large sample size, the empirical distribution should closely resemble the theoretical normal distribution.
#check that the bell shaped curve is symmeterical confirming that its normal distribution
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Probability of success (e.g., getting a head in a coin toss)
p_success = 0.5

# Number of trials (for a single trial)
n_trials = 1

# Generate Bernoulli samples
samples = np.random.binomial(n_trials, p_success, size=1000)

# Plot histogram of the generated samples
plt.hist(samples, bins=[-0.5, 0.5, 1.5], density=True, alpha=0.6, color='b')

# Add labels and title
plt.title('Bernoulli Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.grid(axis='x')

plt.show()
```
