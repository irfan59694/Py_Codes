```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=(15,15),facecolor='w')
ax1=fig.add_axes((0,0,1,1))
x=np.linspace(-10,10,100)
y=np.sin(x)
y1=np.cos(x)
ax1.plot(x,y,color='r',lw=5,ls='--',marker='x',label='sin(x)')
ax1.plot(x,y1,color='g',label='cos(x)')
ax1.set_xlabel('x',fontsize=15,fontfamily='Times New Roman',color='g')
ax1.set_ylabel('sin(x)',fontsize=15,fontfamily='Times New Roman',color='g',rotation=0)
ax1.spines['right'].set_color('None')
ax1.spines['top'].set_color('None')
ax1.spines['bottom'].set_position(('data',0))
ax1.spines['left'].set_position(('data',0))
ax1.set_xticks([-10,-5,5,10])
ax1.legend(loc=3)
ax2=fig.add_axes((0,1,0.1,0.1))
ax2.plot(x,y,color='r',lw=5,ls='--',marker='x',label='sin(x)')
plt.show()
```
